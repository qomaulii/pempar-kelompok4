```python
import matplotlib.pyplot as plt

# ===============================
# DATA HASIL PENGUJIAN
# ===============================

proses = [1, 2, 4, 6, 8]
ukuran_data = [10000, 100000, 500000, 1000000]
ukuran_label = ['10.000', '100.000', '500.000', '1.000.000']

# waktu per proses [10rb, 100rb, 500rb, 1jt]
waktu_per_proses = {
    1: [0.1229, 1.3977, 3.9688, 14.0442],
    2: [0.0804, 1.0117, 2.4882, 10.9784],
    4: [0.0684, 0.7195, 1.8915,  8.0910],
    6: [0.0664, 0.7373, 1.7753,  7.7613],
    8: [0.0723, 0.7292, 1.6022,  7.2893]
}

colors  = ['#E91E63', '#2196F3', '#4CAF50', '#FF9800', '#9C27B0']
markers = ['o', 's', '^', 'D', 'v']

# ===============================
# GRAFIK 1 — Jumlah Proses vs Waktu Eksekusi
# ===============================

data_per_ukuran = {
    '10.000' : [0.1229, 0.0804, 0.0684, 0.0664, 0.0723],
    '100.000': [1.3977, 1.0117, 0.7195, 0.7373, 0.7292],
    '500.000': [3.9688, 2.4882, 1.8915, 1.7753, 1.6022],
    '1.000.000': [14.0442, 10.9784, 8.0910, 7.7613, 7.2893]
}

fig, ax = plt.subplots(figsize=(10, 6))
for i, (label, waktu) in enumerate(data_per_ukuran.items()):
    ax.plot(proses, waktu,
            color=colors[i], marker=markers[i],
            linewidth=2, markersize=7,
            label=f'{label} data')

ax.set_xlabel('Jumlah Proses MPI', fontsize=12)
ax.set_ylabel('Waktu Eksekusi (detik)', fontsize=12)
ax.set_title('Perbandingan Jumlah Proses MPI Terhadap Waktu Eksekusi\nBerdasarkan Ukuran Dataset', fontsize=13, fontweight='bold')
ax.legend(title='Ukuran Data', fontsize=10)
ax.set_xticks(proses)
ax.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('grafik_proses_vs_waktu.png', dpi=150)
plt.show()
print('Grafik 1 tersimpan: grafik_proses_vs_waktu.png')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for i, p in enumerate(proses):
    ax.plot(ukuran_data, waktu_per_proses[p],
            color=colors[i], marker=markers[i],
            linewidth=2, markersize=7,
            label=f'{p} proses')

ax.set_xscale('log')
ax.set_xlabel('Jumlah Data', fontsize=12)
ax.set_ylabel('Waktu Eksekusi (detik)', fontsize=12)
ax.set_title('Perbandingan Ukuran Data Terhadap Waktu Eksekusi\nBerdasarkan Jumlah Proses MPI', fontsize=13, fontweight='bold')
ax.legend(title='Jumlah Proses', fontsize=10)
ax.set_xticks(ukuran_data)
ax.set_xticklabels(ukuran_label)
ax.xaxis.set_minor_locator(plt.NullLocator())  # <-- hapus tick kecil
ax.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('grafik_data_vs_waktu.png', dpi=150)
plt.show()
print('Grafik 2 tersimpan: grafik_data_vs_waktu.png')
```

```python
# ===============================
# DATA LENGKAP (sampai 64 proses)
# ===============================

data_lengkap = {
    '10.000': {
        'proses': [1, 2, 4, 6, 8],
        'waktu' : [0.1229, 0.0804, 0.0684, 0.0664, 0.0723]
    },
    '100.000': {
        'proses': [1, 2, 4, 6, 8, 10, 16, 32, 64],
        'waktu' : [1.3977, 1.0117, 0.7195, 0.7373, 0.7292, 0.7032, 0.6597, 0.6861, 0.7356]
    },
    '500.000': {
        'proses': [1, 2, 4, 6, 8, 10, 16, 32, 64],
        'waktu' : [3.9688, 2.4882, 1.8915, 1.7753, 1.6022, 1.4493, 1.3843, 1.3325, 1.4904]
    },
    '1.000.000': {
        'proses': [1, 2, 4, 6, 8, 10, 16, 32, 64],
        'waktu' : [14.0442, 10.9784, 8.0910, 7.7613, 7.2893, 6.9400, 6.9005, 6.0331, 6.3794]
    }
}

subplot_colors = {
    '10.000'   : '#E91E63',
    '100.000'  : '#2196F3',
    '500.000'  : '#4CAF50',
    '1.000.000': '#FF9800'
}
subplot_markers = {
    '10.000'   : 'o',
    '100.000'  : 's',
    '500.000'  : '^',
    '1.000.000': 'D'
}

for label, d in data_lengkap.items():
    semua_proses = d['proses']
    semua_waktu  = d['waktu']

    # Cari titik optimal dari semua data
    min_idx = semua_waktu.index(min(semua_waktu))
    opt_x   = semua_proses[min_idx]
    opt_y   = semua_waktu[min_idx]

    # Tentukan batas tampilan — satu langkah setelah optimal, minimal sampai 8
    if min_idx + 1 < len(semua_proses):
        batas = semua_proses[min_idx + 1]
    else:
        batas = semua_proses[-1]

    batas = max(8, batas)
    tampil_proses = [p for p in semua_proses if p <= batas]
    tampil_waktu  = semua_waktu[:len(tampil_proses)]

    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(tampil_proses, tampil_waktu,
            color=subplot_colors[label], marker=subplot_markers[label],
            linewidth=2, markersize=7)

    # Tandai titik optimal
    ax.scatter([opt_x], [opt_y], color='red', zorder=5, s=100)
    ax.annotate(f'Optimal\n{opt_x} proses\n{opt_y:.4f}s',
                xy=(opt_x, opt_y),
                xytext=(opt_x + batas * 0.05, opt_y + (max(tampil_waktu) - min(tampil_waktu)) * 0.15),
                arrowprops=dict(arrowstyle='->', color='red', lw=1.5),
                fontsize=9, color='red', fontweight='bold')
    
    ax.set_title(f'Perbandingan Jumlah Proses MPI Terhadap Waktu Eksekusi\nDataset {label} Data', fontsize=12, fontweight='bold')
    ax.set_xlabel('Jumlah Proses MPI', fontsize=11)
    ax.set_ylabel('Waktu Eksekusi (detik)', fontsize=11)
    ax.set_xticks(tampil_proses)
    ax.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.savefig(f'grafik_optimal_{label}.png', dpi=150)
    plt.show()
    print(f'Grafik {label} tersimpan')
```
